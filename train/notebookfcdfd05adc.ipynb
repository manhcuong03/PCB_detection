```python
from ultralytics import YOLO
# Load a model
data_path = '/kaggle/input/pcb-detection/data.yaml'
#'đường dẫn đến file data.yaml trong dataset'
model = YOLO("yolov8n.pt")  # build a new model from YAML

# Train the model
results = model.train(data=data_path, epochs=1000, imgsz=640, device = 0) 
```

```python
from ultralytics import YOLO
import cv2

# Load a model
weight = '/kaggle/working/runs/detect/train2/weights/best.pt'
model = YOLO(weight)

# Predict on the image
results = model.predict('/kaggle/input/pcb-detection/test/images/PCBA_26_jpg.rf.edf8805a5525a41ea7565091fd755249.jpg', show=False, save=True)

for result in results:
    predictions = result.boxes
    
    # Extract bounding boxes, confidence scores, and labels
    boxes = predictions.xyxy.cpu().numpy()  # Bounding boxes in [x1, y1, x2, y2]
    scores = predictions.conf.cpu().numpy()  # Confidence scores
    labels = predictions.cls.cpu().numpy()  # Labels
    
    print("Bounding Boxes:\n", boxes)
    print("Confidence Scores:\n", scores)
    print("Labels:\n", labels)

# Save the result image
result_image_path = "/kaggle/working/runs/predict/image.jpg"  # Adjust based on your output location
output_image = cv2.imread(result_image_path)  # Load the saved prediction result

if output_image is not None:
    cv2.imwrite("output_image.jpg", output_image)
    print("Image saved as output_image.jpg")
else:
    print("No output image found to save.")
```
